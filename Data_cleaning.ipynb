# *Cleaning Data using Pandas*

# Cooking

```python
import pandas as pd
```

```python
data_1=pd.read_csv('/content/Web-scrap_data - cooking_30-40.csv')

data_2=pd.read_csv('/content/Web-scrap_data - cooking_40-50.csv')


data_3=pd.read_csv('/content/cooking_20-30.csv')


data_4=pd.read_csv('/content/cooking_5-20.csv')

# Concatenate them into one DataFrame
data = pd.concat([data_1, data_2, data_3, data_4], ignore_index=True)
```

```python
data=data.drop(['_id','profileType','userId','engagements','picture','isPrivate','isVerified','isHidingLikes','distance','id','name','lat','lon','code','relevance','profileId' ], axis =1)
```

```python
print(data.duplicated())
```

```python
data.drop_duplicates(inplace = True)
```

```python
from google.colab import files
data.to_csv('cleaned_cooking_data.csv')
files.download('cleaned_cooking_data.csv')
```
